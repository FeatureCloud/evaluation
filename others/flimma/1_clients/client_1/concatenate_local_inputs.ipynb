```python
import pandas as pd
import numpy as np
```

```python
k =5
counts = []
design = []
for cli in range(k):
    cli_name = "client_"+str(cli+1)
    df_counts = pd.read_csv("flimma_data/flimma/"+str(k)+"_clients/"+cli_name+"/counts.tsv",sep = "\t",index_col=0)
    print(df_counts.shape[1])
    counts.append(df_counts)

    df_design = pd.read_csv("flimma_data/flimma/"+str(k)+"_clients/"+cli_name+"/design.tsv",sep = "\t",index_col=0)
    # adding binary columns to design to model k-1 batch effects
    for cli2 in range(k):
        if cli2+1!=1:
            cli_name2 = "client_"+str(cli2+1)
            df_design[cli_name2] = 0
            if cli_name2 == cli_name:
                df_design[cli_name2] = 1
    design.append(df_design)
    
counts = pd.concat(counts,axis=1)
design = pd.concat(design,axis=0)

counts.index.name = "gene_name"
counts.to_csv(str(k)+"_clients.counts.tsv",sep= "\t")

design.index.name = "sample_name"
design.to_csv(str(k)+"_clients.design.tsv",sep= "\t")
```
