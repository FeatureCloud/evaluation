```python
import FeatureCloud.api.cli.test.commands as fc
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tests = fc.list(controller_host='http://localhost:8002', format='dataframe')
display(tests)
tests['initializedAt'] = tests['initializedAt'].apply(lambda x: x.replace('T', ' ').split('.')[0] if x is not None else None)
tests['finishedAt'] = pd.to_datetime(tests['finishedAt'])
tests['initializedAt'] = pd.to_datetime(tests['initializedAt'])
tests['createdAt'] = pd.to_datetime(tests['createdAt'])
tests['Runtime (in s)'] = (tests.finishedAt - tests.initializedAt).dt.total_seconds()
tests['Execution time (in s)'] = (tests.finishedAt - tests.createdAt).dt.total_seconds()
tests['Number of clients'] = tests['instancesCount']
tests['Number of iterations'] = tests.dirs.apply(lambda x: x[0].split("/")[-2].split("_")[0])
tests['Number of iterations'] = tests['Number of iterations'].astype(int)
tests = tests.iloc[:, -4:]
tests.to_csv('tests.csv')
tests.tail()
```

```python
sns.color_palette("tab10")
f, ax = plt.subplots(1, 1, figsize=(6, 3), dpi=300)
sns.lineplot(data=tests.reset_index(), y='Runtime (in s)', x='Number of clients', hue='Number of iterations', ax=ax, palette='tab10')
sns.scatterplot(data=tests.reset_index(), y='Runtime (in s)', x='Number of clients', hue='Number of iterations', ax=ax, palette='tab10')
h, l = ax.get_legend_handles_labels()
ax.legend(h[:3], ['1 iteration', '5 iterations', '10 iterations'])
plt.show()
f.tight_layout()
f.savefig('tests.png')
f.savefig('tests.svg')
f.savefig('tests.eps')
```
